# AERQ Satisfaction Survey Results

```python
df_quality = pd.read_csv('./Quality-summary-20210505T100901.csv', sep=';', header=0)
df_quality.shape
df_comm = pd.read_csv('./Communication-summary-20210505T100752.csv', sep=';', header=0)
df_comm.shape
df_trans = pd.read_csv('./Transparency-summary-20210505T101055.csv', sep=';', header=0)
df_trans.shape
df_agile = pd.read_csv('./Applicaton+of+SAFe+Framework-summary-20210505T100837.csv', sep=';', header=0)
df_agile.shape
df_roadmap = pd.read_csv('./Joint+Solution+Roadmap+_+Leveraging+Synergies-summary-20210505T100850.csv', sep=';', header=0)
df_roadmap.shape

```

```python
# add a column with the number of voters per row
q1 = 'What rating do you give the BC Core product that we delivered so far?'
quality_q1_filter = df_quality['Question'] == q1
df_quality_q1 = df_quality.loc[quality_q1_filter, :]
df_quality_q1['#Voters'], df_quality_q1['Percentage'] = df_quality['Current Result'].str.split(',', 1).str
df_quality_q1['#Voters'] = df_quality_q1['#Voters'].str.extract('(\d+)', expand=False).astype(int)
q1_count = df_quality_q1['#Voters'].value_counts()

sns.set(style="darkgrid")
sns.barplot(df_quality_q1.Choices, df_quality_q1['#Voters'], alpha=0.9)
plt.title('What rating do you give the BC Core product that we delivered so far?')
plt.ylabel('Number of Votes', fontsize=12)
plt.xlabel('Ratings', fontsize=12)
plt.xticks(rotation=45)
plt.show()

q2 = 'What rating do you give the others product (System Services, CoTo/DevPortal) that we delivered so far?'
quality_q2_filter = df_quality['Question'] == q2
df_quality_q2 = df_quality.loc[quality_q2_filter, :]
df_quality_q2['#Voters'], df_quality_q2['Percentage'] = df_quality['Current Result'].str.split(',', 1).str
df_quality_q2['#Voters'] = df_quality_q2['#Voters'].str.extract('(\d+)', expand=False).astype(int)
q2_count = df_quality_q2['#Voters'].value_counts()

sns.set(style="darkgrid")
sns.barplot(df_quality_q2.Choices, df_quality_q2['#Voters'], alpha=0.9)
plt.title('What rating do you give the others product (System Services, CoTo/DevPortal) that we delivered so far?')
plt.ylabel('Number of Votes', fontsize=12)
plt.xlabel('Ratings', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
